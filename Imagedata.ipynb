```python
import imageio as img
```

```python
# Read image 
im = img.imread('chest-220.dcm')
```

```python
# Display the image type and shape
print("The image shape : " , type(im))
print("The Shape of the image" , im.shape)
```

```python
# Display the image metadata
print(im.meta.keys())
```

```python
# It is evident that the data is in dictionary format 
# lets print the dictionary items
print(im.meta.items())
```

```python
for key ,value in im.meta.items():
    print('Key:',key, '\t' 'Value:' , value)
    
```

```python
# Create a 3D stack of images
# load images
import numpy as np
#read in the 2D images
im1 = img.imread('chest-221.dcm')
im2= img.imread('chest-222.dcm')
im3= img.imread('chest-224.dcm')
```

```python
# Stack the images vertical
vol=np.stack([im1,im2,im3])
```

```python
print('The shape of the image after stacking them up', vol.shape)
```

```python
# print the metadata of the vol
volume = img.volread('chest')
```

```python
print('Available metadata :' , volume.meta.keys())
```

```python
print('Shape of the image Volume' , volume.shape)
```

```python
# Let me calculate the image field of view
# calculate using the image shape and the image sample resolution
s0 , s1 , s2 = volume.meta.shape
a,b,c=volume.meta['shape']
```

```python
sample1 , sample2 , sample3 = volume.meta['sampling']
```

```python
print(s0,s1,s2)
print(a,b,c)
```

```python
# Calculate the field of view of the image 
print(s0*sample1,s1*sample2,sample3*s2)
```

```python
# Initialize figure and axes grid
fig, axes = plt.subplots(nrows=2, ncols =1)
```

```python
# Draw an image on each subplot
axes[0].imshow(im1, cmap='gray')
axes[1].imshow(im2,cmap='gray')
```

```python
# Remove ticks/labels and render
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```

```python
volume_new= img.volread('sunnybrook-cardiac-mr/SCD2001_000/')

```

```python
# Plot the images on a subplots array 
fig, axes = plt.subplots(nrows =1 , ncols =10,figsize=(200, 200))

# Loop through subplots and draw image
for ii in range(10):
    im = volume_new[ii,:,:]
    axes[ii].imshow(im, cmap='gray')
    axes[ii].axis('off')
    
# Render the figure
plt.show()
```
